```python
import pathlib
import json
country = json.load(open('countries.geo.json'))
```

```python
print(country)
```

```python

m = folium.Map(
    location=[-59.1759, -11.6016],
    tiles='Mapbox Bright',
    zoom_start=1
)
folium.GeoJson(
    country,
    name='geojson'
).add_to(m)


m
```

```python
import pandas as pd
info = pd.read_excel("Book3.xlsx")
```

```python
m = folium.Map()

# Add the color for the chloropleth:
m.choropleth(
 geo_data=country,
 name='choropleth',
 data=info,
 columns=['country_n', 'tc'],
 key_on='feature.id',
 fill_color='YlGn',
 fill_opacity=0.7,
 line_opacity=0.2,
 legend_name='tc (%)'
)
folium.LayerControl().add_to(m)
```
